### 30. 形態素解析結果の読み込み

```python
with open("Outputs/neko.txt.mecab") as in_file:
    neko_morphs = []
    sentence = []
    for line in in_file:
        line = line.rstrip()
        if line == "EOS":
            neko_morphs.append(sentence)
            sentence = []
        else:
            if line[0] != "\u3000":
                line = line.split()
                line[1] = line[1].split(",")
                morph_dict = {}
                morph_dict["surface"] = line[0]
                morph_dict["base"] = line[1][6]
                morph_dict["pos"] = line[1][0]
                morph_dict["pos1"] = line[1][1]
                sentence.append(morph_dict)
    print(neko_morphs[2])
```

### 31. 動詞

```python
for sentence in neko_morphs:
    for word in sentence:
        if word["pos"] == "動詞":
            print(word["surface"], end=", ")
```

### 32. 動詞の原形

```python
bases = set()
for sentence in neko_morphs:
    for word in sentence:
        if word["pos"] == "動詞":
            bases.add(word["base"])

for base in bases:
    print(base, end=", ")
```

### 33. サ変名詞

```python
for sentence in neko_morphs:
    for word in sentence:
        if word["pos"] == "名詞" and word["pos1"] == "サ変接続":
            print(word["base"], end=", ")
```

### 34. 「AのB」

```python
for sentence in neko_morphs:
    for i in range(len(sentence)-2):
        if sentence[i]["pos"] == "名詞" and sentence[i+1]["surface"] == "の" and sentence[i+2]["pos"] == "名詞":
            print(sentence[i]["surface"] + sentence[i+1]["surface"] + sentence[i+2]["surface"], end=", ")
```

### 36. 単語の出現頻度

```python
from collections import Counter
word_cnt = Counter()
for sentence in neko_morphs:
    for word in sentence:
        word_cnt[word["base"]] += 1

#量が多いので100件だけ表示
print(word_cnt.most_common()[:100])
```

### 37. 頻度上位10語

```python
import matplotlib.pyplot as plt
top10 = word_cnt.most_common()[:10]
x = [wordset[0] for wordset in top10]
y = [wordset[1] for wordset in top10]
plt.bar(x, y)
plt.show()
```

### 38. ヒストグラム

```python
#0〜1000の階級が大きすぎて他が表示されないので対数表示
word_frequency = word_cnt.values()
plt.hist(word_frequency, bins=100, log=True)
plt.show()
```

### 39. Zipfの法則

```python
from operator import itemgetter

word_cnt_sort = sorted(word_cnt.items(), key=itemgetter(1))[::-1]
x = [i for i, item in enumerate(word_cnt_sort)]
y = [value for key, value in word_cnt_sort]
plt.plot(x, y)
plt.xscale("log")
plt.yscale("log")
plt.show()
```
